```python
import matplotlib.pyplot as plt
import numpy as np

# Data
durations = ['1s', '2s', '3s', '4s']
baseline_eval_eer = [7.11, 3.39, 2.77, 2.85]
baseline_hidden_eer = [11.77, 10.75, 13.08, 11.87]
multi_view_eval_eer = [6.63, 3.20, 2.72, 2.79]
multi_view_hidden_eer = [10.50, 10.21, 10.29, 10.50]

x = np.arange(len(durations))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, baseline_eval_eer, width, label='Baseline Eval EER')
rects2 = ax.bar(x + width/2, multi_view_eval_eer, width, label='Multi-View Eval EER')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('EER (%)')
ax.set_title('EER by Model and Duration')
ax.set_xticks(x)
ax.set_xticklabels(durations)
ax.legend()

# Adding values on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
