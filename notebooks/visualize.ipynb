```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data
durations = ['1s', '2s', '3s', '4s']
baseline_eval_eer = [7.11, 3.39, 2.77, 2.85]
baseline_hidden_eer = [11.77, 10.75, 13.08, 11.87]
multi_view_eval_eer = [6.19, 2.77, 2.31, 2.29]
multi_view_hidden_eer = [10.06, 8.74, 8.54, 8.89]

x = np.arange(len(durations))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 6))

# Plotting the bars
rects1 = ax.bar(x - 1.5 * width, baseline_eval_eer, width, label='Baseline Eval EER')
rects2 = ax.bar(x - 0.5 * width, baseline_hidden_eer, width, label='Baseline Hidden EER')
rects3 = ax.bar(x + 0.5 * width, multi_view_eval_eer, width, label='Multi-View Eval EER')
rects4 = ax.bar(x + 1.5 * width, multi_view_hidden_eer, width, label='Multi-View Hidden EER')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('EER (%)')
ax.set_title('EER by Model, Type, and Duration')
ax.set_xticks(x)
ax.set_xticklabels(durations)
ax.legend()

# Adding values on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

# Overlay line plot to show trends
ax.plot(x, baseline_eval_eer, marker='o', color='blue', linestyle='dashed', linewidth=2, markersize=6, label='Baseline Eval EER Trend')
ax.plot(x, baseline_hidden_eer, marker='o', color='orange', linestyle='dashed', linewidth=2, markersize=6, label='Baseline Hidden EER Trend')
ax.plot(x, multi_view_eval_eer, marker='o', color='green', linestyle='dashed', linewidth=2, markersize=6, label='Multi-View Eval EER Trend')
ax.plot(x, multi_view_hidden_eer, marker='o', color='red', linestyle='dashed', linewidth=2, markersize=6, label='Multi-View Hidden EER Trend')

fig.tight_layout()

plt.show()

# Statistical Summary
data = {
    'Duration': durations,
    'Baseline Eval EER': baseline_eval_eer,
    'Baseline Hidden EER': baseline_hidden_eer,
    'Multi-View Eval EER': multi_view_eval_eer,
    'Multi-View Hidden EER': multi_view_hidden_eer
}

df = pd.DataFrame(data)
summary = df.describe().loc[['mean', 'std']]
print(summary)
```
