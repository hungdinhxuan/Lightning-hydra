```python
import pandas as pd
from eval_metrics_DF import compute_eer

aasist_scores = "/nvme2/hungdx/Lightning-hydra/logs/results/ieee_paper_benchmark/xlsr_aasist_mdt_var/ASVSpoof21_Eval_paper_ieee_spl_2026_xlsr_aasist_mdt_xlsr_aasist_mdt_var.txt"

scores = pd.read_csv(aasist_scores, sep=' ', header=None)

scores.columns = ['utt_id', 'spoof_score', 'bonafide_score']


metadata = pd.read_csv('/nvme2/hungdx/Lightning-hydra/data/TakHemlata SSL_Anti-spoofing_protocols/keys/metadata.csv')

merge_scores = pd.merge(scores, metadata, on='utt_id', how='left')



```

```python
aasist_eval_scores = merge_scores[merge_scores['subset'] == 'eval']

aasist_hidden_scores = merge_scores[merge_scores['subset'] == 'hidden']
```

```python
# calculate eer for aasist_eval_scores
bona_cm_eval = aasist_eval_scores[aasist_eval_scores['label'] == 'bonafide']['bonafide_score'].values
spoof_cm_eval = aasist_eval_scores[aasist_eval_scores['label'] == 'spoof']['bonafide_score'].values
eer_cm_eval, th_cm_eval = compute_eer(bona_cm_eval, spoof_cm_eval)
print(f"EER for aasist_eval_scores: {eer_cm_eval * 100}%")



bona_cm_hidden = aasist_hidden_scores[aasist_hidden_scores['label'] == 'bonafide']['bonafide_score'].values
spoof_cm_hidden = aasist_hidden_scores[aasist_hidden_scores['label'] == 'spoof']['bonafide_score'].values
eer_cm_hidden, th_cm_hidden = compute_eer(bona_cm_hidden, spoof_cm_hidden)
print(f"EER for aasist_hidden_scores: {eer_cm_hidden * 100}%")


```
